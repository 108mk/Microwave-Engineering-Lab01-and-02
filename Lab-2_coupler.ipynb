```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#case-I
```

```python
S21_coupling = pd.read_csv('S21_coupling.csv', sep=';')
```

```python
S21_coupling
```

```python
#Mean of |S12| and |S21|
S12_mean=S21_coupling["S12-Magnitude (dB)"].mean()
S21_mean=S21_coupling["S21-Magnitude (dB)"].mean()
```

```python
S12_mean
```

```python
S21_mean
```

```python
# plot of S vs f; x-axis in log scale.
S21_coupling.plot(x='Frequency', y=['S12-Magnitude (dB)', 'S21-Magnitude (dB)' ], figsize=(10,5), grid=True)
plt.xscale("log")

S21_coupling.plot(x='Frequency', y=['S12-Phase (°)', 'S21-Phase (°)' ], figsize=(10,5), grid=True)
plt.xscale("log")
```

```python
# plot of S vs f; x-axis in linear scale
S21_coupling.plot(x='Frequency', y=['S12-Magnitude (dB)', 'S21-Magnitude (dB)'], figsize=(10,5), grid=True)

S21_coupling.plot(x='Frequency', y=['S12-Phase (°)','S21-Phase (°)' ], figsize=(10,5), grid=True)
```

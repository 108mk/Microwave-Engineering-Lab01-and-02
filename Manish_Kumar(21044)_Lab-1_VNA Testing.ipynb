## Part-III: Determination of Wire length using Phase change

```python
wire_length = pd.read_csv('wire_length.csv', sep=';')
```

```python
wire_length
```

```python
# plot of S-Phase vs frequency; x-axis in LOG SCALE

wire_length.plot(x='Frequency', y=['S21-Phase (°)'], figsize=(10,5), grid=True)
#plt.xscale("log")
plt.axvline(x=650000000, color='magenta', linestyle='--')
plt.axvline(x=1300000000, color='green', linestyle='--')
plt.xlabel("Frequency")
plt.ylabel("Phase($^\circ$)")
```

# Conclusion: 
> $ w_1= 0.65\times 10^{9}Hz, \ w_2 = 1.3 \times 10^{9} Hz \implies dw =0.65\times 10^{9}Hz $ 
### Reference plane is shifted to port-1, wave travels through the length of wire and acquire a phase change of 180 degree. 
### We can now calculate the length of wire as- 
> $ wt=\phi  \\  \Delta w \times t = \Delta \phi \\ t = \frac{L}{c/n}\\ \implies L = \frac{\Delta \phi (c/n)}{(\Delta w)}$
### Taking refractive index of dielectric material (PFTE) n = 1.4 gives:
>> $L=10.36 \ cm $

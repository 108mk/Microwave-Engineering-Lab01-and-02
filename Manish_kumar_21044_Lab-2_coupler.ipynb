# Case-I: Measurement of coupling factor
### The configuration is shown below. 
## "All other ports are matched."
### Coupling factor = $ |S_{21}|\ in\ dB\ = 20 \log_{10} \frac{V_2 ^-}{V_1 ^+}$  

```python
S21_coupling = pd.read_csv('S21_coupling.csv', sep=';')
```

```python
S21_coupling
```

```python
# We will restrict our calculation till the frequency of 2 GHz. Or,
S21_coupling.loc[[429]]
```

```python
Coupling_factor = S21_coupling["S21-Magnitude (dB)"][0:429].mean()
```

```python
Coupling_factor
```

```python
# plot of S vs f; x-axis in log scale.
S21_coupling.plot(x='Frequency', y=['S21-Magnitude (dB)' ], figsize=(10,5), grid=True)
plt.xscale("log")
plt.xlabel("Frequency (Hz)")
plt.ylabel("$ |S_{21}| $")
```
